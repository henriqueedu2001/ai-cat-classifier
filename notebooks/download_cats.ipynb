```python
import os
import yaml
from bing_image_downloader import downloader
```

```python
# important folder
SRC_DIR = os.path.abspath('./../src/')
CONF_DIR = os.path.join(SRC_DIR, 'conf/')
DATA_DIR = os.path.join(SRC_DIR, 'data/')

SRC_DIR, CONF_DIR, DATA_DIR
```

```python
cat_classes = []
cat_classes_conf_path = os.path.join(CONF_DIR, 'cat_classes.yaml')

# reads the cat classes from the conf file
with open(cat_classes_conf_path, 'r') as conf_file:
    content = yaml.safe_load(conf_file)
    cat_classes = content['cat_classes']

print(cat_classes)
```

```python
cat_views = []
cat_views_conf_path = os.path.join(CONF_DIR, 'cat_views.yaml')

# reads the cat classes from the conf file
with open(cat_views_conf_path, 'r') as conf_file:
    content = yaml.safe_load(conf_file)
    cat_views = content['cat_views']

print(cat_views)
```

```python
cat_image_types = []

# mix the views and the classes to create different types
for cat_view in cat_views:
    for cat_class in cat_classes:
        cat_image_type = f'{cat_class} cat breed {cat_view}'
        cat_image_types.append(cat_image_type)

print(cat_image_types)
```

```python
# creating paths for the images
imgs_paths = {cat_image_type: os.path.join(DATA_DIR, cat_image_type) for cat_image_type in cat_image_types}

# creating the folders for every image type
for cat_type, path in imgs_paths.items():
    os.mkdir(path)
    
print(imgs_paths)
```

```python
# downloads all the images from bing
for cat_type, path in imgs_paths.items():
    downloader.download(
        cat_type,
        limit=100,
        output_dir=path,
        adult_filter_off=True,
        force_replace=False,
        timeout=5*60,
        verbose=True
    )
```
